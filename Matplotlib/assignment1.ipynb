```python
import matplotlib.pyplot as plt
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
kohli = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
rohit = [600, 700, 1800, 1700, 1300, 1900, 700, 1500, 1000, 500]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
```

```python
plt.plot(years,kohli,label="Virat Kohli",color= "red", linestyle ='-.',linewidth =3 )
plt.plot(years,rohit,'g^',label = "Rohit Sharma", color= "blue",linestyle ='-.',linewidth =2)
plt.plot(years,sehwag,'ro--' ,label="Virender Sehwag",color= "green",linestyle ='--',linewidth=1)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title(" 10 years of hypothetical runs.")
plt.legend()
```
