```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")  # Choose any style you like
```

```python
# Data
labels = ["Dhoni", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels)
plt.show()
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs, labels=labels, colors=colors)
plt.show()
```

```python
plt.pie(runs,labels=labels,colors=colors,wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '--'})
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor': 'red', 'linewidth': 2, 'linestyle': '-.','fill':False}
)
plt.show()
```

```python
explode = [0.1, 0, 0, 0]  # Only highlight Sachin's slice

plt.pie(runs, labels=labels, explode=explode, colors=colors)
plt.title("Exploded Pie Example")
plt.show()
```

```python
explode = [0.1, 0, 0, 0]  # Only highlight Sachin's slice
# for 3d view
plt.pie(runs, labels=labels, explode=explode, colors=colors,shadow= True) 
plt.title("Exploded Pie Example")
plt.show()
```

```python
explode= [0.1,0,0,0]

plt.pie(runs,labels=labels,explode =explode, colors=colors,shadow= True,startangle=90)
plt.title("Rotated pe chart example")
plt.show()
```

```python
explode= [0.1,0,0,0]

plt.pie(runs,labels=labels,explode= explode,colors= colors,shadow=True,autopct= '%1.1f%%')
plt.title("chart with percentage")
plt.show()
```

```python
# Too many categories
# not to use for much data
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust', 'Swift', 'PHP']
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1]

plt.pie(usage, labels=languages, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage (Crowded Example)")
plt.show()
```
