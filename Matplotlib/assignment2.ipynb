```python
import matplotlib.pyplot as plt
import numpy as np
years = [1990, 1992, 1994, 1996, 1998]
sachin = [500, 700, 1100, 1500, 1800 ]
sehwag = [0, 200, 900, 1400, 1600 ]
kohli  = [0, 300, 500, 800, 1100 ]
```

```python
width = 0.25  # Thickness of each bar
y = np.arange(len(years))  # Positions for grouping

# Horizontal bars
plt.barh(y - width, sachin, height=width, color="skyblue", label="Sachin Tendulkar")
plt.barh(y, sehwag, height=width, color="red", label="Virender Sehwag")
plt.barh(y + width, kohli, height=width, color="green", label="Virat Kohli")

# Labels and formatting
plt.xlabel("Total Runs in First 5 Years")
plt.ylabel("Year")
plt.yticks(y, years)
plt.title("First 5-Year Performance of Indian Batsmen")
plt.legend()
plt.tight_layout()

# Show
plt.show()
```
